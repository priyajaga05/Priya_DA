```python
# # Import NumPy for numerical operations (arrays, argmax, etc.)
import numpy as np

# Import matplotlib to visualize images
import matplotlib.pyplot as plt

# Import the MNIST handwritten digits dataset from Keras
from keras.datasets import mnist

# Import Sequential model (layers added one after another)
from keras.models import Sequential

# Import Dense (fully connected layer) and Flatten (2D → 1D)
from keras.layers import Dense, Flatten


# -----------------------------
# Load the MNIST dataset
# -----------------------------

# Loads handwritten digit images and their labels
# x_train, y_train → training images and labels
# x_test, y_test   → testing images and labels
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Convert pixel values from int to float and normalize them
# Original pixel range: 0–255
# After normalization: 0.0–1.0

x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255


# -----------------------------

# Create a Sequential model
model = Sequential()

# Flatten 28x28 image into a 1D vector of size 784
model.add(Flatten(input_shape=(28, 28)))

# Add a hidden Dense layer with 128 neurons and ReLU activation
# ReLU helps the model learn non-linear patterns
model.add(Dense(128, activation='relu'))

# Add output layer with 10 neurons (digits 0–9)
# Softmax converts outputs into probabilities
model.add(Dense(10, activation='softmax'))

# Compile the model
# -----------------------------

# optimizer='adam' → updates weights efficiently
# loss='sparse_categorical_crossentropy' → used when labels are integers (0–9)
# metrics=['accuracy'] → track accuracy during training
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)


# -----------------------------
# Train the model
# -----------------------------

# Train the model using training data
# epochs=5 → the model sees the entire dataset 5 times
model.fit(x_train, y_train, epochs=5)

# Evaluate the model
# -----------------------------

# Test the model on unseen test data
# Returns loss and accuracy
test_loss, test_acc = model.evaluate(x_test, y_test)

# Print the test accuracy
print(f'Test accuracy: {test_acc}')


# -----------------------------
# Make predictions
# -----------------------------

# Predict probabilities for each test image
# Output shape: (number_of_images, 10)
predictions = model.predict(x_test)


# -----------------------------
# Display prediction for first test image
# -----------------------------

# Get the index of the highest probability (predicted digit)
predicted_label = np.argmax(predictions[0])

# Print the predicted digit
print(f'Predicted label for the first test image: {predicted_label}')


# -----------------------------
# Visualize the first test image
# -----------------------------

# Display the first test image in grayscale
plt.imshow(x_test[0], cmap='gray')

# Show the predicted label as the title
plt.title(f'Predicted label: {predicted_label}')

# Render the image
plt.show()

```
